```python
print(statistics.variance(ages))  #using statistics find sample variance
```

```python
# find variance manually
def variance(data):
    n=len(ages)
    #mean
    mean=sum(data)/n
    #variance
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/n  #Population Variance
    return variance
```

```python
print(variance(ages))
```

```python
# find variance manually
def variance(data):
    n=len(ages)
    #mean
    mean=sum(data)/n
    #variance
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-1)  #Sample Variance
    return variance
```

```python
print(variance(ages))
```

```python
# find variance manually
def variance(data,dof=0):  #degree of freedom for population =0,for sample=1
    n=len(ages)
    #mean
    mean=sum(data)/n
    #variance
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-dof)  #Population Variance
    return variance
```

```python
print(variance(ages))
```

```python
# find variance manually
def variance(data,dof=1):  #degree of freedom for population =0,for sample=1
    n=len(ages)
    #mean
    mean=sum(data)/n
    #variance
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-dof)  #Population Variance
    return variance
```

```python
print(variance(ages))
```

# Check whether a distribution is Normal or Gaussian Distribution or not...

```python
# If you want to check whether a feature is gaussian or normal distributed
# Use Q-Q Plot
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))  #Size of  line chart
    plt.subplot(1,2,1)  # 1 row ,2 columns,in one plot
    sns.histplot(sample)
    plt.subplot(1,2,2) #In Two  plots
    stat.probplot(sample,dist='norm',plot=pylab) # proplot probability plot,x-axis=sample,y-axis=distribution normal
    plt.show()
```

```python
plot_data(s)
```

we can see in the plot both sample and normal distribution line are straight that means the data 's' is normal distributed.

```python
plot_data(a)
```

In this plot we can see the both lines are not straight that means the data 'a' is not normal distributed.

Let's we check for iris data column=petal width.

```python
plot_data(df['sepal_width'])
```

Both are in same line , so data is normal distributed.
